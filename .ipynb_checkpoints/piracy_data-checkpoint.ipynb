```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import scipy
import seaborn as sns
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
df1 = pd.read_csv('./ytsyifytorrent-movies/movies_metadata.csv')

#       Importing Dataset and Libarys
```

```python
# Droping columns and only keeping column in the category list
category = ['budget','genres','imdb_id','original_language','original_title','popularity','release_date','revenue','status','vote_average','vote_count','title']
df = df1[category]
#df.head()
```

```python
# Renaming some of the columns for simplicity
header = {'imdb_id':'imdb','original_language':'language','original_title':'og_title','spoken_languages':'sp_languages'}
df = df.rename(columns = header)
```

```python
# Graphing a Bar ploting Movie Revenue by Movie Rating
fig = plt.figure(figsize=(7.5,6))
plt.style.use('fivethirtyeight')
ax=plt.axes()


plt.bar(df.vote_average, df.revenue)
plt.title(' Movie Revenue by Movie Rating')
plt.xlabel("Movie Rating")
plt.ylabel("Movie Revenue(Billions)")
plt.show()
```

```python
# Movie Budget by movie Ratings 
fig = plt.figure(figsize=(7.5,6))
plt.style.use('fivethirtyeight')
ax=plt.axes()

plt.bar(df.vote_average,df.budget )
plt.title('Movie Budget By Movie Ratings')
plt.xlabel("Movie Rating")
plt.ylabel("Movie budget(100 million)")

plt.show()
```

```python
# Graphing Movie Budget by movie renenue
fig = plt.figure(figsize=(7.5,6))
plt.style.use('fivethirtyeight')
ax=plt.axes()

plt.scatter(df.budget, df.revenue)
plt.title('Movie Budget By Movie Revenue')
plt.xlabel("Movie Budget(100 millions)")
plt.ylabel("Movie Revenue(Billions)")
plt.show()
```
