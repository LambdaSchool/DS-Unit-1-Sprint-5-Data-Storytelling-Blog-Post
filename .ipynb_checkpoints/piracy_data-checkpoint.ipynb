```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import scipy
import seaborn as sns
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

# fix duplicate titles look for relation in title and og_title
# replace ??? on og title 
# integtate imbd data set
# 1 merge both dataset with imdb id
# budget and revinew 0 skewing the data set. 

df1 = pd.read_csv('./ytsyifytorrent-movies/movies_metadata.csv')
```

```python
category = ['budget','genres','imdb_id','original_language','original_title','popularity','release_date','revenue','status','vote_average','vote_count','title']
df = df1[category]
#df.head()
```

```python
header = {'imdb_id':'imdb','original_language':'language','original_title':'og_title','spoken_languages':'sp_languages'}
df = df.rename(columns = header)
df.budget.value_counts()
```

```python
vote_a = np.array(df.vote_average)
vote_c = np.array(df.vote_count)
pop = np.array(df.popularity)
poprev = df.loc[(df['revenue'] > 100), ['popularity','revenue']]
budrev = df.loc[(df['budget'] > 100 ) & (df['revenue'] > 100), ['budget','revenue']]

budrev2 = budrev.loc[((budrev['budget'] < 120994180.77867822 ) | (budrev['budget'] > 133892302.13807796)) & ((budrev['revenue'] < 120994180.77) | (budrev['revenue'] > 133892302.13)), ['budget','revenue']]
ci, _, _ = scipy.stats.bayes_mvs(budrev.revenue, alpha=.99)
ci
```

```python
fig = plt.figure(figsize=(7.5,6))
plt.style.use('fivethirtyeight')
ax=plt.axes()
plt.scatter(budrev2.budget, budrev2.revenue, marker='v', color='b')
plt.title('Movie Budget By Movie Revenue')
plt.show()
```
