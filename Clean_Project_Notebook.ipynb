## Data.

World Bank Suicide data:

```
suicide = pd.read_csv(io.BytesIO(uploaded['API_SH.STA.SUIC.P5_DS2_en_csv_v2_10517940.csv']), skiprows=4)

print(suicide.shape)

suicide.head()
```

**GDP per Capita PPP**

```
uploaded = files.upload()
```

```
ppp = pd.read_csv(io.BytesIO(uploaded['API_NY.GDP.PCAP.PP.CD_DS2_en_csv_v2_10515265.csv']), skiprows=4)

print(ppp.shape)

ppp.head()
```

```
suicide = suicide.rename(columns = {'Country': 'Country Name'})
```

```
df_ppp = pd.merge(suicide, ppp, how='inner', on='Country Name')

print(df_ppp.shape)

df_ppp.head(25)
```

```
df_ppp = df_ppp[['Country Name', 'Indicator Name_x', '2016_x', 'Indicator Name_y', '2016_y']]

df_ppp.shape
```

```
df_ppp.isna().sum()
```

```
df_ppp = df_ppp.dropna()

print(df_ppp.shape)

df_ppp.head()
```

```
df_ppp = df_ppp.rename(columns = {'2016_x': 'Suicide Rate',
                                 '2016_y': 'GDP per Capita, PPP'})
```

```
sns.lmplot(x='GDP per Capita, PPP', y='Suicide Rate', data=df_ppp);

# GDP per Capita is a little bit more confused but there is a trend.
# So it is not totally the fault of the World Happiness Report. There does seem to be at least a slight trend of increased gdp per capita and increased suicide rates.
```
