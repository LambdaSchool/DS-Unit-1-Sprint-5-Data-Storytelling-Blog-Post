## Data.

World Bank Suicide data:

```
uploaded = files.upload()
```

```
suicide = pd.read_csv(io.BytesIO(uploaded['API_SH.STA.SUIC.P5_DS2_en_csv_v2_10517940.csv']), skiprows=4)

print(suicide.shape)

suicide.head()
```

```
suicide = suicide.drop(columns=['Indicator Code', 'Country Code', 'Unnamed: 63', '2017', '2018'])

suicide.head()
```

```
suicide = suicide.rename(columns = {'Country Name': 'Country'})

suicide.head()
```

World Happiness Report data:

```
uploaded = files.upload()
```

```
hreport = pd.read_csv(io.BytesIO(uploaded['2016.csv']))

print(hreport.shape)

hreport.head()
```

Making main dataframe:

```
df = pd.merge(hreport, suicide, how='left', on='Country')

print(df.shape)

df.head()
```

```
df = df[['Country', 'Happiness Score', 'Lower Confidence Interval', 
        'Upper Confidence Interval', 'Economy (GDP per Capita)', 'Family', 
        'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 
        'Generosity', 'Dystopia Residual', '2016']]

df = df.rename(columns = {'2016': 'Suicide Rate'})

print(df.shape)

df.head()
```

```
df.head(200)
```

```
df = df.set_value('44', 'Suicide Rate', 9.19)
```

```
sns.lmplot(x="Happiness Score", y="Suicide Rate", data=df);
```
