```python
import pandas as pd

def csv_combiner(csvs):
    combined_csv = pd.read_csv(csvs[0], index_col=0)
    for i in range(1, len(csvs)):
        combined_csv = combined_csv.append(pd.read_csv(csvs[i], index_col=0))
    return combined_csv

Antrim_1901_csvs = [
'1901_Antrim_Ardclinis.csv',
'1901_Antrim_Ballyclug.csv',
'1901_Antrim_Broughshane.csv',
'1901_Antrim_Carncastle.csv',
'1901_Antrim_Newtown_Crommelin.csv',
'1901_Antrim_Glenarm.csv',
'1901_Antrim_Glencloy.csv',
'1901_Antrim_Glenravel.csv',
'1901_Antrim_Glenwherry.csv',
'1901_Antrim_Longmore.csv',
'1901_Antrim_Red_Bay.csv',
'1901_Antrim_Slemish.csv'
]

Down_1901_csvs = [
'1901_Down_Ballykeel.csv',
'1901_Down_Bryansford.csv',
'1901_Down_Greencastle.csv',
'1901_Down_Kilkeel.csv',
'1901_Down_Killowen.csv',
'1901_Down_Maghera.csv',
'1901_Down_Mourne_Park.csv',
'1901_Down_Mullartown.csv',
'1901_Down_Rosstrevor.csv',
'1901_Down_Fofanny.csv'
]

Fermanagh_Tyrone_1901_csvs = [
'1901_Fermanagh_Ballinamallard.csv',
'1901_Fermanagh_Ballycassidy.csv',
'1901_Fermanagh_Ballydoolagh.csv',
'1901_Fermanagh_Ballyreagh.csv',
'1901_Fermanagh_Castle_Coole.csv',
'1901_Fermanagh_Clabby.csv',
'1901_Fermanagh_Imeroo.csv',
'1901_Fermanagh_Irvinestown.csv',
'1901_Fermanagh_Lisbellaw.csv',
'1901_Fermanagh_Lisnarick.csv',
'1901_Fermanagh_Magheraculmoney.csv',
'1901_Fermanagh_Milltown.csv',
'1901_Fermanagh_Newporton.csv',
'1901_Fermanagh_Rockfield_or_Killadeas.csv',
'1901_Fermanagh_Tempo.csv',
'1901_Tyrone_Carryglass.csv',
'1901_Tyrone_Dromore.csv',
'1901_Tyrone_Drumharvey.csv',
'1901_Tyrone_Kilskeery.csv',
'1901_Tyrone_Moorfield.csv',
'1901_Tyrone_Trillick.csv',
'1901_Tyrone_Tullyclunagh.csv'
]

Antrim_1911_csvs = [
'1911_Antrim_Ardclinis.csv',
'1911_Antrim_Ballyclug.csv',
'1911_Antrim_Broughshane.csv',
'1911_Antrim_Carncastle.csv',
'1911_Antrim_Glenarm.csv',
'1911_Antrim_Glencloy.csv',
'1911_Antrim_Glenravel.csv',
'1911_Antrim_Glenwherry.csv',
'1911_Antrim_Longmore.csv',
'1911_Antrim_Newtown_Crommelin.csv',
'1911_Antrim_Red_Bay.csv',
'1911_Antrim_Slemish.csv'
]

Down_1911_csvs = [
'1911_Down_Ballykeel.csv',
'1911_Down_Bryansford.csv',
'1911_Down_Fofanny.csv',
'1911_Down_Greencastle.csv',
'1911_Down_Kilkeel.csv',
'1911_Down_Killowen.csv',
'1911_Down_Maghera.csv',
'1911_Down_Mourne_Park.csv',
'1911_Down_Mullartown.csv',
'1911_Down_Newcastle.csv',
'1911_Down_Rosstrevor.csv'
]

Fermanagh_Tyrone_1911_csvs = [
'1911_Fermanagh_Ballycassidy.csv',
'1911_Fermanagh_Ballydoolagh.csv',
'1911_Fermanagh_Ballyreagh.csv',
'1911_Fermanagh_Ballinamallard.csv',
'1911_Fermanagh_Castle_Coole.csv',
'1911_Fermanagh_Clabby.csv',
'1911_Fermanagh_Imeroo.csv',
'1911_Fermanagh_Irvinestown.csv',
'1911_Fermanagh_Lisbellaw.csv',
'1911_Fermanagh_Lisnarick.csv',
'1911_Fermanagh_Magheraculmoney.csv',
'1911_Fermanagh_Milltown.csv',
'1911_Fermanagh_Newporton.csv',
'1911_Fermanagh_Rockfield_or_Killadeas.csv',
'1911_Fermanagh_Tempo.csv',
'1911_Tyrone_Carryglass.csv',
'1911_Tyrone_Dromore.csv',
'1911_Tyrone_Drumharvey.csv',
'1911_Tyrone_Kilskeery.csv',
'1911_Tyrone_Moorfield.csv',
'1911_Tyrone_Trillick.csv',
'1911_Tyrone_Tullyclunagh.csv'
]

Antrim_1901 = csv_combiner(Antrim_1901_csvs).sort_values(by=[
    'house_id', 'resident_number']).reset_index(drop='True')
Down_1901 = csv_combiner(Down_1901_csvs).sort_values(by=[
    'house_id', 'resident_number']).reset_index(drop='True')
Fermanagh_Tyrone_1901 = csv_combiner(Fermanagh_Tyrone_1901_csvs).sort_values(by=[
    'house_id', 'resident_number']).reset_index(drop='True')

Antrim_1911 = csv_combiner(Antrim_1911_csvs).sort_values(by=[
    'house_id', 'resident_number']).reset_index(drop='True')
Down_1911 = csv_combiner(Down_1911_csvs).sort_values(by=[
    'house_id', 'resident_number']).reset_index(drop='True')
Fermanagh_Tyrone_1911 = csv_combiner(Fermanagh_Tyrone_1911_csvs).sort_values(by=[
    'house_id', 'resident_number']).reset_index(drop='True')
```

```python
census_1911 = Antrim_1911.append(Down_1911).append(Fermanagh_Tyrone_1911)
```

```python
roman_catholic = ['R Catholic', 'Catholic', 'R C', 'Roman Catholic Church',
                  'Catholic Church', 'R Catholic Church', 'Roman Catholick',
                  'R C Church', 'R Catholice', 'Rom Catholic', 
                  'Catholic Roman', 'R Catholick', 'Roman C Church',
                  'Roman Cath']

church_of_ireland = ['Irish Church', 'Ch of Ireland', 'Church of England', 
                     'Episcopalian', 'Church Ireland', 
                     'Protestant Episcopalian', 'Episcopal Church',
                     'Episcopal Church of Ireland']

presbyterian = ['Presbyterian Church', 'Presbytarian', 'Presbeterian',
                'Prespyterian', 'Reformed Presbyterian', 'Prespeterian',
                'Presbyterians', 'Presbyterian Ga', 'Pres', 'Presbyterian Ch']

census_1911['religion'].replace(
    to_replace=roman_catholic,
    value='Roman Catholic',
    inplace=True
)

census_1911['religion'].replace(
    to_replace=church_of_ireland,
    value='Church of Ireland',
    inplace=True
)

census_1911['religion'].replace(
    to_replace=presbyterian,
    value='Presbyterian',
    inplace=True
)
```

```python
major_religion_census_1911 = census_1911[(census_1911['religion']=='Roman Catholic') |
            (census_1911['religion']=='Presbyterian') | 
            (census_1911['religion']=='Church of Ireland')]
```
