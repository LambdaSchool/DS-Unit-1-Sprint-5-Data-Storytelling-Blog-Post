```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```
#Load datasets
air= pd.read_csv('https://raw.githubusercontent.com/alvinwalker314/Datasets/master/world_war_2_aircrafts.csv')
sea= pd.read_csv('https://raw.githubusercontent.com/alvinwalker314/Datasets/master/WW2%20Ships%20-%20Sheet1.csv')
vehicles= pd.read_csv('https://raw.githubusercontent.com/alvinwalker314/Datasets/master/WW2_Vehicles.csv')
land= pd.read_csv('https://raw.githubusercontent.com/alvinwalker314/Datasets/master/infantry%20-%20Sheet1.csv')
```

```
#filled and renames User of aircraft
air['Operator(s) if different from country of origin'].fillna(air['Country of Origin'], inplace= True)
air.rename(columns={'Operator(s) if different from country of origin':'User of Aircraft'}, inplace= True)
air.dropna(inplace= True)
```

```
#values to plot Airplanes & Missiles
prototypes=[80, 39, 40, 30, 37, 15, 12, 21]
bombers=   [24, 10, 11, 9, 4, 8, 5, 4]
fighters=  [22, 14, 12, 13, 10, 12, 5, 9]
transport= [21, 9, 13, 10, 5, 6, 0, 2]
seaplane=  [7, 6, 4, 10, 3, 2, 5, 1]
training=  [16, 2, 8, 5, 5, 0, 2, 11]
misc=      [11, 1, 0, 2, 2, 0, 0, 0]
recon=     [5, 10, 1, 3, 1, 1, 1, 0]

#set size of graph
fig=plt.figure(figsize=(20, 10))

#set x axis locations
x= 8
locations= np.arange(x)
plt.xticks(locations, ('US', 'Germany', 'UK', 'Japan', 'Italy', 'Russia', 'France', 'Other'));

#Plot values
p1= plt.bar(locations, prototypes)
p2= plt.bar(locations, bombers, bottom= prototypes)
p3= plt.bar(locations, fighters, bottom= np.array(prototypes)+ np.array(bombers))
p4= plt.bar(locations, transport, bottom=  np.array(prototypes)+ np.array(bombers)+ np.array(fighters))
p5= plt.bar(locations, seaplane, bottom= np.array(prototypes)+ np.array(bombers)+ np.array(fighters)+np.array(transport))
p6= plt.bar(locations, training, bottom= np.array(prototypes)+ np.array(bombers)+ np.array(fighters)+np.array(transport)+ np.array(seaplane))
p7= plt.bar(locations, recon, bottom= np.array(prototypes)+ np.array(bombers)+ np.array(fighters)+np.array(transport)+ np.array(seaplane)+ np.array(training))
p8= plt.bar(locations, misc, bottom= np.array(prototypes)+ np.array(bombers)+ np.array(fighters)+np.array(transport)+ np.array(seaplane)+ np.array(training)+ np.array(recon))

#set legend
plt.legend((p1[0], p2[0], p3[0], p4[0], p5[0], p6[0], p7[0], p8[0]), ('Prototypes', 'Bombers', 'Fighters', 'Transport', 'Seaplanes', 'Training', 'Recon', 'Misc'), fontsize=15)

#titles
plt.title('WW2 Aircraft and Missile Development', fontsize= 25)
plt.xlabel('Country of Origin', fontsize=20)
plt.ylabel('Total Models Developed', fontsize=20);
plt.grid()
```
