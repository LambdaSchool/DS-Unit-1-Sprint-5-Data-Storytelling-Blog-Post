```
# Creating function that finds max or min value in specified range, and the row
def get_max_min(start, end, direction):

  date_range = df[(df.index >= start) & (df.index <= end)]
  if direction == 'min':
    max_or_min_value = date_range['Real Price'].min()
  elif direction == 'max':
    max_or_min_value = date_range['Real Price'].max()
  else:
    print("Must enter 'min' or 'max'")
  
  return df[df['Real Price'] == max_or_min_value]

get_max_min(dt.datetime(1871, 8, 1), dt.datetime(1906, 9, 1), 'min')
```
