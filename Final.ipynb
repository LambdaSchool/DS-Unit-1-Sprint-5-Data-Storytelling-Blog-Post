```python
def get_genre(tags):
    masterlist = []
    for tag in tags:
        for index in range(len(aggregate_genres)):
            for genre, subgenre_list in aggregate_genres[index].items():
                for subgenre in subgenre_list:
                    if tag in subgenre:
                        masterlist.append(genre)
    if not masterlist:
        return (np.nan)
    else:
        #print (masterlist)
        m = max(masterlist,key=masterlist.count)
        return (m)
```

```python
top_20['tags'] = top_20['tags'].apply(lambda x:get_genre(x))
```
