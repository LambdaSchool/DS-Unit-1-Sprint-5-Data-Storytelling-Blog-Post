```python
# Let's take a look at lobbying expenditures on an annualized basis

year_range = np.arange(1999, 2020)
```

```python
no_duplicate_ID_df = bigbertha.drop_duplicates(subset='ID')
```

```python
no_duplicate_ID_df.shape
```

```python
for i in (year_range):
    annualized_descriptive_stats = pd.DataFrame(no_duplicate_ID_df.groupby(by='Year').describe())
    
for i in (year_range):
    annualized_expenditure_counts = no_duplicate_ID_df.groupby(by='Year').count()
    
for i in (year_range):
    annualized_avg_expenditure = no_duplicate_ID_df.groupby(by='Year').mean()
    
for i in (year_range):
    annualized_med_expenditure = no_duplicate_ID_df.groupby(by='Year').median()
```

```python
annualized_descriptive_stats
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-whitegrid')
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
filing_counts = pd.DataFrame(annualized_expenditure_counts['ID'])
filing_counts.reset_index
filing_counts['Year']=year_range
print (filing_counts.columns)
filing_counts.columns = ['ID', 'Year']
filing_counts.reset_index(drop=True)
filing_counts
```

```python
dims = (12, 7)
fig, ax = plt.subplots(figsize=dims)
sns.set_style('dark')
sns.light_palette("navy")
filing_count_plot = sns.barplot(x=filing_counts.Year,
           y=filing_counts.ID,color='r')
plt.xticks(rotation=40, fontsize=15)
plt.yticks(fontsize=15)
ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed')
filing_count_plot.set_xlabel("",fontsize=1.5)
filing_count_plot.set_ylabel("",fontsize=1.5)
filing_count_plot.axes.set_title("Number of Lobbying Disclosures Filed by Year", fontsize=30)
filing_count_plot.set(xlim=(0, 21));
```
