### Comparing career length in years based on draft round

```python
# Binning career length into five year increments
cut_points = [0, 5, 10, 15, 20]
label_names = ["0-5", "6-10", "11-15", "16-20"]
draft_combine_data['career_len_bins'] = pd.cut(draft_combine_data['Career_Len'], cut_points, labels=label_names)
draft_combine_data['career_len_bins'].head()
```

```python
draft_combine_data['career_len_bins'].value_counts()
```

```python
crosstab_table = pd.crosstab(draft_combine_data['Rnd'], draft_combine_data['career_len_bins'])
crosstab_table
```

```python
crosstab_table_pct = pd.crosstab(draft_combine_data['Rnd'], draft_combine_data['career_len_bins'], margins=True,
                                normalize='index')
crosstab_table_pct
```
