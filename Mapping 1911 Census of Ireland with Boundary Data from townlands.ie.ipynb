```python
townland_to_map_df = census_1911_df.groupby(['townland_to_map_url']).size().reset_index(name='population')

townland_to_map_df = census_1911_df[
    ['townland_to_map', 'townland_to_map_url']].merge(
    townland_to_map_df, on=['townland_to_map_url']).drop_duplicates()

gjson_to_gpd = gpd.read_file('townland_boundaries_from_townlands_ie.geojson')
area_df = pd.DataFrame({'townland_to_map_url': pd.Series(gjson_to_gpd['T_IE_URL']), 'area': gjson_to_gpd.area})

townland_to_map_df = townland_to_map_df.merge(
    area_df, on=['townland_to_map_url'])

townland_to_map_df['area'] = (townland_to_map_df['area'] / 1000000)

townland_to_map_df['population_density'] = (
    townland_to_map_df['population'] / townland_to_map_df['area'])
```

```python
marital_status_data = census_1911_df[
    ((census_1911_df['marital_status'] == 'Married') |
     (census_1911_df['marital_status'] == 'Widowed')) &
    (census_1911_df['age'] < 70)]

marital_status_ct = pd.crosstab(marital_status_data['townland_to_map_url'], 
                 marital_status_data['marital_status']).reset_index()

marital_status_ct[
    'pct_wid_of_mar_and_wid_under_age_70'] = (marital_status_ct['Widowed'] / (
    marital_status_ct['Widowed'] + marital_status_ct['Married']))

townland_to_map_df = townland_to_map_df.merge(
    marital_status_ct[['townland_to_map_url', 
                      'pct_wid_of_mar_and_wid_under_age_70']],
    on='townland_to_map_url', how='outer')
```

```python
religion_ct = pd.crosstab(census_1911_df['townland_to_map_url'], 
                 census_1911_df['religion']).reset_index()

townland_to_map_df = townland_to_map_df.merge(
    religion_ct[['townland_to_map_url', 'Roman Catholic']],
    on='townland_to_map_url', how='outer')

townland_to_map_df['pct_catholic'] = (
    townland_to_map_df['Roman Catholic'] / townland_to_map_df['population'])

townland_to_map_df = townland_to_map_df.drop(columns=['Roman Catholic'])
```

```python
literacy_data = census_1911_df[census_1911_df['age'] > 8]

literacy_ct = pd.crosstab(literacy_data['townland_to_map_url'], 
                 literacy_data['literacy']).reset_index()

townland_to_map_df = townland_to_map_df.merge(
    literacy_ct[['townland_to_map_url', 'Cannot read']],
    on='townland_to_map_url', how='outer')

townland_to_map_df['pct_illiterate_above_age_8'] = (
    townland_to_map_df['Cannot read'] / townland_to_map_df['population'])

townland_to_map_df = townland_to_map_df.drop(columns=['Cannot read'])
```

```python
irish_language_ct = pd.crosstab(census_1911_df['townland_to_map_url'], 
                 census_1911_df['irish_language']).reset_index()

irish_language_ct['irish_spoken'] = (irish_language_ct['Irish and English'] + 
                       irish_language_ct['Irish'])

townland_to_map_df = townland_to_map_df.merge(
    irish_language_ct[['townland_to_map_url', 'irish_spoken']],
    on='townland_to_map_url', how='outer')

townland_to_map_df['pct_speaks_irish'] = (
    townland_to_map_df['irish_spoken'] / townland_to_map_df['population'])

townland_to_map_df = townland_to_map_df.drop(columns=['irish_spoken'])
```

```python
children_born_data = census_1911_df[census_1911_df['children_born'] > 0]

children_living_df = children_born_data.groupby([
    'townland_to_map_url'])[[
    'children_born', 'children_living']].sum().reset_index()

children_living_df['pct_children_deceased'] = (
    (children_living_df['children_born'] - 
     children_living_df['children_living']) / 
    children_living_df['children_born'])

townland_to_map_df = townland_to_map_df.merge(
    children_living_df[['townland_to_map_url', 
                      'pct_children_deceased']],
    on='townland_to_map_url', how='outer')
```

```python
boundaries = folium.features.GeoJson(gjson_to_gpd)

def embed_map(m):
    from IPython.display import HTML

    m.save('index.html')
    with open('index.html') as f:
        html = f.read()

    iframe = '<iframe srcdoc="{srcdoc}" style="width: 100%; height: 750px; border: none"></iframe>'
    srcdoc = html.replace('"', '&quot;')
    return HTML(iframe.format(srcdoc=srcdoc))
```

```python
bins = list(townland_to_map_df[
    townland_to_map_df['population_density'] > 0][
    'population_density'].quantile(
    [(1/5)*i for i in range(6)]))
bins.insert(0, 0)

# for mourne area: location=[54.1450, -6.04043]
# for slemish area: location=[54.9530892, -6.043359]
# for ballinamallard area: location=[54.4287082, -7.5471696]
density_map = folium.Map(location=[54.1450, -6.04043], zoom_start=11, 
                        tiles='cartodbpositron')

density_map.add_child(boundaries)

density_map.choropleth(
    geo_data=gjson_to_gpd,
    name='choropleth',
    data=townland_to_map_df,
    columns=['townland_to_map_url', 'population_density'],
    key_on='feature.properties.T_IE_URL',
    fill_color='PuBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='population_density',
    bins=bins,
    reset=True
)
folium.LayerControl().add_to(density_map)

embed_map(density_map)
```
