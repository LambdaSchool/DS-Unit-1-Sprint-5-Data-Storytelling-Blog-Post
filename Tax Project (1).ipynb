```python
graph1['Death Rate'] = df['Death Rate']
```

```python
graph1['Unemployment Rate'] = df[' Unemployment rate %']
```

```python
graph1 = graph1.rename({'Death Rate': 'Death Rate Per 1000 People'}, axis=1)
```

```python
f, ax = plt.subplots(figsize=(20, 12))
sns.set(style="whitegrid")
sns.despine(f, left=True, bottom=True)
sns.scatterplot(y="GDP Per Capita", x="Tax-Rate",
                size = "Death Rate Per 1000 People",
                hue = 'Unemployment Rate',
                palette=sns.cubehelix_palette(dark=.1, light=.8, as_cmap=True, hue=1, rot=-.4),
                sizes=(50, 700), linewidth=0,
                data=graph1, ax=ax)
plt.legend(fontsize='large')
```

```python
ax = pd.cut(graph1['Tax-Rate'], bins=3).value_counts().plot.bar(rot=0, title='Total Countries For Each Tax Rate', colormap='summer')
ax.set_xlabel("Tax Rates by %")
ax.set_ylabel("Number of Countries")
```
